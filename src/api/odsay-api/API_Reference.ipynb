# API Response Format

## 최상위 노드

### result
- **Type**: 확장노드
- **Required**: Y
- **Occurrences**: 1
- **Description**: 데이터를 포함하는 최 상위 노드

## result 노드의 하위 요소들

### 1. searchType
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 결과 구분 (도시내 or 도시간 직통 or 도시간 환승)
  - 0: 도시내
  - 1: 도시간 직통
  - 2: 도시간 환승

### 2. outTrafficCheck
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 도시간 "직통" 탐색 결과 유무(환승 X)
  - 0: False
  - 1: True

### 3. busCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 버스 결과 개수

### 4. subwayCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 지하철 결과 개수

### 5. subwayBusCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: “버스+지하철” 결과 개수

### 6. pointDistance
- **Type**: double
- **Required**: Y
- **Occurrences**: 1
- **Description**: 출발지(SX, SY)와 도착지(EX, EY)의 직선 거리 (미터)

### 7. startRadius
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 출발지 반경

### 8. endRadius
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 도착지 반경

### 9. path
- **Type**: 확장노드
- **Required**: Y
- **Occurrences**: 1...n
- **Description**: 결과 리스트 확장 노드

## path 노드의 하위 요소들

### 1. pathType
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 결과 종류
  - 1: 지하철
  - 2: 버스
  - 3: 버스+지하철

### 2. info
- **Type**: 확장노드
- **Required**: Y
- **Occurrences**: 1
- **Description**: 요약 정보 확장 노드

## info 노드의 하위 요소들

### 3. totalTime
- **Type**: int
- **Required**: Y
- **Occurrences**: 1

### 4. payment

### 5. busTransitCount

### 6. subwayTransitCount

### 7. mapObj

### 8. firstStartStation

- **Required**: Y
- **Occurrences**: 1
- **Description**: 최초 출발역/정류장

### 9. firstStartStationKor

### 10. firstStartStationJpnKata

### 11. lastEndStation

### 12. lastEndStationKor
- **Type**: string
- **Required**: Y
- **Occurrences**: 1
- **Description**: 최종 도착역/정류장 국문 (다국어 서비스 시 표출)

### 13. lastEndStationJpnKata
- **Type**: string
- **Required**: Y
- **Occurrences**: 1
- **Description**: 최종 도착역/정류장 일문(가타카나) (lang = 2 인 경우 표출)

### 14. totalStationCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 총 정류장 합

### 15. busStationCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 버스 정류장 합

### 16. subwayStationCount
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 지하철 정류장 합

### 17. totalDistance
- **Type**: double
- **Required**: Y
- **Occurrences**: 1
- **Description**: 총 거리

### 18. checkIntervalTime
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 배차간격 체크 기준 시간(분)

### 19. checkIntervalTimeOverYn
- **Type**: string
- **Required**: Y
- **Occurrences**: 1
- **Description**: 배차간격 체크 기준시간을 초과하는 노선이 존재하는지 여부 (Y/N)

### 20. totalIntervalTime
- **Type**: int
- **Required**: Y
- **Occurrences**: 1
- **Description**: 전체 배차간격 시간(분)

## path 노드의 하위 요소들

### 1. subPath
- **Type**: 확장노드
- **Required**: Y
- **Occurrences**: 1...n
- **Description**: 이동 교통 수단 정보 확장 노드

  - 1: 지하철
  - 2: 버스
  - 3: 도보

### 2. distance
- **Type**: double
- **Required**: Y
- **Occurrences**: 1
- **Description**: 이동 거리
